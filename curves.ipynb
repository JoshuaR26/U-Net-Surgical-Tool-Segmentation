```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a Pandas DataFrame
df = pd.read_csv('./log/log_3_NEW.csv')

# Extracting data from DataFrame
epoch = df['epoch']
train_jaccard_distance = df['jaccard_distance']
train_loss = df['loss']
val_jaccard_distance = df['val_jaccard_distance']
val_loss = df['val_loss']

# Plotting
plt.figure(figsize=(10, 5))

plt.plot(epoch, train_jaccard_distance, label='Train Jaccard Distance')
plt.plot(epoch, train_loss, label='Train Loss')
plt.plot(epoch, val_jaccard_distance, label='Validation Jaccard Distance')
plt.plot(epoch, val_loss, label='Validation Loss')

plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Training and Validation Metrics Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
